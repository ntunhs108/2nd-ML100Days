* 基本上，有三種方法來解決一個多標籤分類問題
     (http://202.119.32.195/cache/11/03/lamda.nju.edu.cn/f5353e662bc504485df6291f8f834a47/xuzg.pdf)
     
  1.問題轉換:
    問題轉換方法的核心是“改造樣本數據使其適應現有學習算法”。
    (a)二元關聯（Binary Relevance）:把每個標籤當作單獨的一個類分類問題
    (b)分類器鏈（Classifier Chains）:在分類器鏈中，把4個標籤的問題將被轉換成4種不同的標籤問題(https://cdn-images-1.medium.com/max/640/0*JXVJevCooeJE_IJX.png)
    (c)標籤Powerset（Label Powerset）:標籤powerset將這個問題轉換為一個單一的多類問題(https://cdn-images-1.medium.com/max/640/0*_b6JbmR051nMfIqI.png)
  2.改編算法:
    算法適應方法的核心是“改造現有的單標記學習算法使其適應多標記數據”。
    改編算法是直接執行多標籤分類，而不是將問題轉化為不同的問題子集。
    例如:直接用SKlearn之MLkNN(kNN的多標籤版本)、隨機森林（Random Forest）和嶺回歸（Ridge regression）計算
  3.集成方法:Scikit-Multilearn庫提供不同的組合分類功能，你可以使用它來獲得更好的結果。
    對於直接實現，你可以查看：http://scikit.ml/api/classify.html#ensemble-approaches

其他多標籤分類
* 音頻分類
  我們知道歌曲會被分類為不同的流派。他們也被分類為，如“放鬆的平靜”，或“悲傷的孤獨”等等情感或情緒的基礎。
  來源：http://lpis.csd.auth.gr/publications/tsoumakas-ismir08.pdf
* 圖像分類
  使用圖像的多標籤分類也有廣泛的應用。圖像可以被標記為不同的對象、人或概念。
* 生物信息學
  多標籤分類在生物信息學領域有很多用途，例如，在酵母數據集中的基因分類。
  它還被用來使用幾個未標記的蛋白質來預測蛋白質的多重功能。
* 文本分類
  谷歌新聞所做的是，將每條新聞都標記為一個或多個類別，這樣它就會顯示在不同的類別之下。

* 多標籤分類(Multilabel classification): 每個資料樣本同時具有很多目標標籤(目標變量，target variable).
                                       且這些標籤的屬性不是互相排斥
                                       (例如:一個水果是蘋果又是梨子就是互相排斥；如果文章的話題是宗教、
                                        政治、金融，可同時存在，這就不會互相排斥而形成多標籤)
  註:多類分類(multi-class classification)：資料具有多種類別需要分類，但每一個樣本只屬於一個類別(label)     區別: 
     對於多類分類，最後一層使用softmax函數進行預測，訓練階段使用categorical_crossentropy作為損失函數 
     對於多標籤分類，最後一層使用sigmoid函數進行預測，訓練階段使用binary_crossentropy作為損失函數
