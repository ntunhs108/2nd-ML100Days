```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
unique_house_type = pd.unique(app_train['HOUSETYPE_MODE'])

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):  #len(unique_house_type)=4，i=0~3，不包含4
    plt.subplot(nrows, ncols, i+1)
    
    #A histogram is a representation of the distribution of data.
    #https://blog.csdn.net/brucewong0516/article/details/82494090
    #依布林值選取True的row，並選擇'AMT_CREDIT'那個column
    app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]),'AMT_CREDIT'].hist() 
    plt.title(str(unique_house_type[i]))
plt.show()    
```
