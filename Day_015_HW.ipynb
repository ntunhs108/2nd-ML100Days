```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
numpy.random.random_sample(size=None)
-->Return random floats in the half-open interval [0.0, 1.0).

2*numpy.random.random_sample(size=None)-1
-->Return random floats in the half-open interval [-1.0, 1.0).
"""
```

```python
#用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣
matrix = 2 * np.random.random_sample((10,10)) - 1
matrix
```

```python
plt.figure(figsize=(10,10))

sns.heatmap(matrix)
plt.title("Try to create sample heapmap")
plt.show()

plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)
plt.title("Try to create sample heapmap")
plt.show()
```

```python
plt.figure(figsize=(10,10))

#heatmap輸入值可以是array也可以是dataframe
sns.heatmap(pd.DataFrame(matrix), cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)
plt.title("Try to create sample heapmap")
plt.show()
```

```python
#用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣
nrow = 1000
ncol = 3

matrix = 2 * np.random.random_sample((nrow,ncol)) - 1
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

"""
繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

"""
繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
Please replace "..." to correct plot function
"""
grid.map_upper(sns.scatterplot, alpha = 0.2)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
